Revenue Trends

```python
import matplotlib.pyplot as plt
import numpy as np

revenue_by_subject = df.groupby('subject')['revenue'].sum().sort_values(ascending=False)

# Generate a different color for each segment
colors = plt.cm.tab20.colors[:len(revenue_by_subject)]

plt.figure(figsize=(6,6))  # Reduced size
plt.pie(
    revenue_by_subject.values,
    labels=revenue_by_subject.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors
)
plt.title('Total Revenue Distribution by Subject')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

Top 10 Earning Courses

```python
top_courses = df.sort_values(by='revenue', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='revenue', y='course_title', data=top_courses, palette='crest')
plt.title('Top 10 Revenue-Generating Courses')
plt.xlabel('Revenue')
plt.ylabel('Course Title')
plt.tight_layout()
plt.show()
```

Average Rating per Subject

```python
avg_rating = df.groupby('subject')['Rating'].mean().sort_values()

plt.figure(figsize=(10,6))
sns.barplot(x=avg_rating.values, y=avg_rating.index, palette='coolwarm')
plt.title('Average Course Rating by Subject')
plt.xlabel('Average Rating')
plt.ylabel('Subject')
plt.tight_layout()
plt.show()
```

 Subscriber Count by Course Level

```python
subs_by_level = df.groupby('level')['num_subscribers'].sum().sort_values()

plt.figure(figsize=(8,5))
sns.barplot(x=subs_by_level.values, y=subs_by_level.index, palette='magma')
plt.title('Total Subscribers by Course Level')
plt.xlabel('Number of Subscribers')
plt.ylabel('Course Level')
plt.tight_layout()
plt.show()
```
